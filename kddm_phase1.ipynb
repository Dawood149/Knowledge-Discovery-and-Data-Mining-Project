```python
contracts = pd.read_pickle("contracts_competition.pkl")

identity0 = pd.read_csv("identity_card_0.tsv", sep="\t")

identity1 = pd.read_csv("identity_card_1.csv")

performance = pd.read_csv("performance.tsv", sep="\t")

medical = pd.read_excel("medical_information.xlsx", skiprows=1)

scouting = pd.read_csv("hockey_scouting_notes.csv")

moms = pd.read_json("moms_notes.json")
```

```python
dataframes = {
    "contracts": contracts,
    "identity0": identity0,
    "identity1": identity1,
    "medical": medical,
    "performance": performance,
    "scouting": scouting,
    "moms":moms,
}
```

```python
audit=[]

for name, df in dataframes.items():
    rows=df.shape[0]
    cols=df.shape[1]

    duplicate_rows=df.duplicated().sum()

    total_missing=df.isnull().sum().sum()

    key_cols=[col for col in df.columns if "id" in col.lower()]

    audit.append({
        "file_name": name,
        "rows": rows,
        "columns": cols,
        "duplicate_rows": duplicate_rows,
        "total_missing_values": total_missing,
        "possible_keys": ", ".join(key_cols)

    })

audit_df=pd.DataFrame(audit)
audit_df
```
